```python
# Importing needed libraries
import os, shutil

# Defining folders paths to be read
path = 'D:\\Google Fotos\\'
#nova_path = path
os.chdir(path)

# Generating list of internal folders
pastas = os.listdir() 
print('Pastas:', pastas, '\n') 

# Creating variable to count amount of accumulated files
contador = 0
```

```python
# Loop for organizing photos into folders by year
# Ignore folders that do not have a date format in their name

for pasta in pastas:
    fullpath = path + pasta + '\\Takeout\\Google Fotos\\' 
    if os.path.isdir(fullpath): 
        print('pasta:', pasta)
        n_pastas = 0
        n_arquivos = 0
        
        # Generates daily folder list
        items = os.listdir(fullpath) 
       
        # Checks whether the folder name is a date
        for item in items:
            try: 
                ano = int(item[0:4])
            except:
                continue

            if ano >= 2010 & ano <= 2020:
                # Updates internal folder count
                n_pastas += 1
                
                # Create folder for the year
                dir = os.path.join("D:\\","FOTOS",str(ano))
                if not os.path.exists(dir):
                    os.mkdir(dir)

                # Generates complete file list of the folder
                fullpath = path + pasta + '\\Takeout\\Google Fotos\\' + item + '\\'
                os.chdir(fullpath)
                arquivos = os.listdir()
                
                
                # Defines path
                os.chdir(fullpath)
                lista_arquivos = list()
                
                # If the file is json, discard. If not, move to folder defined in variable dir
                for arquivo in arquivos:
                    if arquivo.endswith('json'):
                        continue 
                    else: 
                        lista_arquivos.append(arquivo)
                        #shutil.copy(arquivo,dir)   
                        #shutil.move(arquivo,dir)
                
                # Updates file counter and accumulated variables
                n_arquivos += len(lista_arquivos)
                contador += len(lista_arquivos) 
        print('Qtde de arquivos:', n_arquivos)    
        print('Qtde de pastas diarias:', n_pastas)
        print('Quantidade acumulada:', contador, '\n') 
    else:
        print(pasta, 'não é uma pasta', '\n')
print('Quantidade total:', contador, '\n')
```
